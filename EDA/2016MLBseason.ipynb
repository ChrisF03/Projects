## Take Me Out To The Ballgame

In this section, we will take a few different looks at attendance and climate in the 2016 season. <br>

```python
venue_attend = df[df['season']=='regular season'].groupby(['venue','home_team'])['attendance'].mean().reset_index()
venue_attend.drop(labels=10,axis=0,inplace=True)
venue_attend.head()
```

```python
fig1 = px.bar(venue_attend, x="venue", y="attendance", color='home_team')
fig1.update_layout(title='Average Attendance per MLB Stadium', xaxis_title='Stadium', yaxis_title='Attendance',showlegend=False,height=600)
fig1
```

Here we can see which MLB stadiums had the highest average attendance throughout the season. Dodger Stadium, home of the Los Angeles Dodgers, took the lead. Averaging nearly 46,000 fans per game, the Dodgers had the highest team payroll for the 2016 season and were a very popular pick to win the World Series that year. Busch Stadium, home of the St.Louis Cardinals, and Rogers Centre, home of the Toronto Blue Jays came in second and third place respectively. 

```python
attendance = df[df['season']=='regular season'].groupby('date')['attendance'].mean().reset_index()
```

```python
fig2 = go.Figure(data=go.Scatter(x=attendance['date'], y=attendance['attendance'], mode='lines', marker=dict(color='purple')))
fig2.update_layout(title='Average Attendance Throughout 2016 MLB Regular Season', xaxis_title='Date', yaxis_title='Attendance')
fig2.show()
```

<p> Attendance during the regular season was highest during Opening Weekend, peaking at an average at almost 42,000 spectators per ballpark on April 4th before nosediving over the next couple of days. Attendance, predictably, fluctuated throughtout the weeks, with the highest points points being during weekends and the lowest during weekdays. The months of June and, especially, July saw a relatively more consistent stream of ticket sales with higher lows being had across the league, this could be attributed to kids and college students having more free time during the summer months as well as the MLB All-Star festivities which are always held in mid-July. The playoff picture starts to become clearer around this time of year as well and the trade deadline, which is usually set for the final day of that month, can also be exciting for fans as they look to see any new players acquired by their teams.

```python
Day = df[df['season']=='regular season'].groupby('day_of_week')['attendance'].mean().reset_index()
week = [ 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
Day['day_of_week'] = pd.Categorical(Day['day_of_week'],categories=week,ordered=True)
Day = Day.sort_values('day_of_week')
Day
```

```python
fig3 = px.line(Day, x="day_of_week", y="attendance", title='Average Attendance per day during regular season')
fig3.update_layout(xaxis_title='Day')
fig3
```

Here can see how attendance varied for each weekday. The week starts out with low attendance across the board and gradually picks up as the days go by with Saturday, predictably, seeing the highest attendance. Sunday, understandably, followed as the day with the second highest amount of seat-fillers as most sunday games are played earlier in the day inviting fans to a relaxing day at the ballpark before starting the work or schoolweek. 

```python
GameType = df[df['season']=='regular season'].groupby('game_type')['attendance'].mean().reset_index()
GameType
```

```python
fig4 = px.bar(GameType, x="game_type", y="attendance",color='game_type')
fig4.update_layout(title='Average Attendance per Game Type', xaxis_title='Type', yaxis_title='Attendance',showlegend=False)
fig4
```

Here we have a closer look at the point made in the previous graph. Games played during the day which usually take place during the weeekends, see an average of nearly 3,000 more fans than night games. 

```python
Sky = df[df['season']=='regular season'].groupby('sky')['attendance'].mean().reset_index()
Sky
```

```python
fig5 = px.bar(Sky, x='sky', y='attendance',color='sky')
fig5.update_layout(title='Average Attendance per Sky Condition',showlegend=False)
fig5
```

Fans showed out more on sunny days and came out almost just as much when it was drizzling, which I find interesting as I though cloudy or maybe even overcast days would see more fans.

## Hey Batter, Batter ...

### Runs by venue 

In this section we will take a closer look at the scoring data provided for venues and teams. 

```python
Runs = df[df['season']=='regular season'].groupby(['venue','home_team'])['home_team_runs','away_team_runs'].sum().reset_index()
Runs.head()
```

```python
Runs['total runs'] = Runs['home_team_runs'] + Runs['away_team_runs']
Runs.drop(labels=10,axis=0,inplace=True)
```

```python
fig6 = px.box(Runs,x='total runs')
fig6.update_layout(title='Distrubtion of Total Runs Scored Per Stadium')
fig6
```

The box plot above gives us a better look at how total runs were distributed by ballpark. The median was 713.5 runs per venue, the lower whisker went down to 604 and the upper whisker went up to 904. There is one outlier stadium where 985 runs were scored at that season. 

```python
fig7 = px.bar(Runs,x='venue',y='total runs',color='home_team',height=600)
fig7.update_layout(title='Total Runs Scored At Each Stadium',showlegend=False)
fig7
```

Here we can see that the stadium that saw the least amount of runs was Marlins Park, home of the Miami Marlins, which does have a reputation for having pitcher-friendly dimensions. The venue setting the upper whisker was Chase Field, home of the Arizona Diamondbacks. The outlier venue was Coors Field, home of the Colorado Rockies in Denver. Coors Field has earned a reputation as a very hitter-friendly ballpark, due to the effect of Denver's high elevation and semi-arid climate on the distances of batted balls. 

### Runs by team 

Now lets take a look at runs and hits for each team ... 

```python
TeamRuns = df[df['season']=='regular season'].groupby(['home_team'])['home_team_runs'].sum().reset_index()
TeamRuns.head()
```

```python
TeamRuns1 = df[df['season']=='regular season'].groupby(['away_team'])['away_team_runs'].sum().reset_index()
TeamRuns1.head()
```

```python
TeamRuns['Runs as away'] = TeamRuns1['away_team_runs']
TeamRuns['Total Runs'] = TeamRuns['Runs as away'] + TeamRuns['home_team_runs']
TeamRuns.rename(columns = {'home_team':'Team', 'home_team_runs':'Runs as home'},inplace=True)
TeamRuns.head()
```

```python
fig8 = px.box(TeamRuns,x='Total Runs')
fig8.update_layout(title='Distrubtion of Total Runs Scored Per Team')
fig8
```

According to the above box plot, the median number of runs scored by team was 723. The highest scoring team put a total of 878 runs on the scoreboard throughout the season, with the lowest scoring team managing a meager 610. 

```python
fig9= go.Figure(data = go.Scatter(x=TeamRuns['Runs as home'], y=TeamRuns['Team'], marker=dict(color="blue", size=12), mode="markers", name="Home"))
fig9.add_trace(go.Scatter(x=TeamRuns['Runs as away'], y=TeamRuns['Team'], marker=dict(color="red", size=12), mode="markers", name="Away"))
fig9.update_layout(title="Runs Scored At Home vs Away", xaxis_title="Runs Scored", yaxis_title="Team", height = 1100)
fig9.update_yaxes(autorange="reversed")
fig9
```

The Colorado Rockies scored the most runs (508) at their home ballpark, Coors Field. They were clearly aided by the high elevation and thin air provided by its location near the Rocky Mountains as Colorado also was amongst the lowest scorers away from home, managing a paltry 337 runs in enemy territory. The Boston Red Sox were the second highest scorers at home, Fenway Park, scoring 477 runs. The Chicago Cubs and the St.Louis Cardinals were the highest scorers away from their respective home fields, scoring 419 and 424 respectively and both scoring more away than at home. 

```python
fig9 = go.Figure(data=[go.Bar(name='Away', x=TeamRuns['Team'], y=TeamRuns['Runs as away'],marker_color='red'), go.Bar(name='Home', x=TeamRuns['Team'], y=TeamRuns['Runs as home'],marker_color='blue', text=TeamRuns['Total Runs'],textposition='outside')])
fig9.update_layout(title='Total Runs Scored per Team', barmode='stack', xaxis_title='Team',yaxis_title='Runs', height=600)
fig9.show()
```

Here we have another look at run production among the 30 teams. The Boston Red Sox scored the most runs overall with a total of 878. They where followed by the Colorado Rockies and the Chicago Cubs in second and third place respectively. 

### Hits by team

Now a look at team hits ... 

```python
Hits = df[df['season']=='regular season'].groupby('home_team')['home_team_hits'].sum().reset_index()
Hits.head()
```

```python
Hits1 = df[df['season']=='regular season'].groupby('away_team')['away_team_hits'].sum().reset_index()
Hits1.head()
```

```python
Hits['Hits away'] =  Hits1['away_team_hits']
Hits['Total hits'] = Hits['home_team_hits'] + Hits['Hits away']
Hits.rename(columns = {'home_team':'Team', 'home_team_hits':'Hits at home'},inplace=True)
Hits.head()
```

```python
fig10 = px.box(Hits,x='Total hits')
fig10.update_layout(title='Distribution of Total Hits by Team')
fig10
```

The median amount of hits for MLB teams in 2016 was 1,409.5. The lowest was 1,275 and the highest was an outlier team who recorded 1,598 hits. 

```python
fig11 = go.Figure(data=[go.Bar(name='Away', x=Hits['Team'], y=Hits['Hits away'],marker_color='orange'), go.Bar(name='Home', x=Hits['Team'], y=Hits['Hits at home'],marker_color='purple', text=Hits['Total hits'],textposition='outside')])
fig11.update_layout(title='Total Hits Recorded per Team', barmode='stack', xaxis_title='Team',yaxis_title='Hits', height=600)
fig11.show()
```

Once again, the Boston Red Sox lead the pack with 1598 hits, followed by, again, the Colorado Rockies and the Arizona Diamondbacks with 1,544 and 1,479 respectively. 

```python
Hits['Runs Scored'] = TeamRuns['Total Runs']
Hits.head()
```

```python
fig12 = px.scatter(Hits, x="Runs Scored", y="Total hits", size='Total hits', color='Team', title="Correlation between Hits and Runs",height=800)
fig12
```

The scatter plot gives us an interesting look at how runs and hits are related. While our intuition might tell us more hits equals more runs, according to this graph, that isn't necessarily true. The relationship between the two variables, while linear and positive, isn't as strong as one might first guess. Hit-producing juggernauts, Colorado and Boston, are in a league of their own in the top right section of the graph but the Miami Marlins, for one, had MORE hits (1,460) but LESS runs scored (655) than a grand total of 21 teams out of 30. The San Diego Padres, on the other hand, had the least amount of hits overall (1,275) in the league but still managed to outscore a total of 9 teams.  <p> Teams with very similar amounts of hits also had much larger differences in runs scored, the biggest example of this being the Atlanta Braves and the Chicago Cubs. These two ballclubs were only seperated by a mere 5 hits, with Atlanta coming in at 1,404 and Chicago at 1,409. Chicago, however, outscored the Atlanta Braves by an impressive 159 runs with a total of 808, while the Braves only managed 649, this perhaps highlighting the importance of good baserunning and in-game strategy. 

## Play Ball ! 

In this section we'll take a look at wins per teams. 

```python
Wins = df[df['season']=='regular season'].groupby('home_team')['home_team_win'].sum().reset_index()
Wins.head()
```

```python
Wins1 = df[df['season']=='regular season'].groupby('away_team')['home_team_loss'].sum().reset_index()
Wins1.head()
```

```python
Wins['home_team_win'] = Wins['home_team_win'] + Wins1['home_team_loss']
Wins.rename(columns = {'home_team':'Team', 'home_team_win':'Total Wins'},inplace=True)
Wins.head()
```

```python
WinsS = Wins.sort_values(["Total Wins"], ascending=True).reset_index()
```

```python
fig13 = px.bar(WinsS,x='Team',y='Total Wins',color='Total Wins',height=600)
fig13.update_layout(title='Total Wins per Team',showlegend=False)
fig13
```

The storied Chicago Cubs were the most winningest ballclub of the 2016 MLB season with a total of 103 games won, followed by the Washington(D.C) Nationals and the Texas Rangers. Interestingly, neither of the highest run producers, Colorado and Boston, were in the top 3. Boston, however, did come in 5th and Colorado just 20th. 

```python
Wins['Total Runs'] = TeamRuns['Total Runs'] 
Wins['Total hits'] = Hits['Total hits']
Wins.head()
```

```python
fig14 = px.scatter(Wins, x="Total Wins", y="Total Runs",size="Total Wins", color="Team",title="Correlation between Wins and Runs Scored",height=800)
fig14.update_layout(xaxis_title='Wins',yaxis_title='Runs Scored',showlegend=True)
fig14
```

Runs did have a positive linear relationship to wins although not a very strong one. The Boston Red Sox scored 70 more runs than the Chicago Cubs but Chicago still won 10 more games than Boston. Many teams with varying amounts of runs scored still wound up with the same or very similar wins, such as the Atlanta Braves (649 runs), Tampa Bay Rays (672), San Diego Padres (686), and the Cincinnati Reds (716) which all ended the season with 68 wins. The Oakland Athletics (653) and the Arizona Diamondbacks (752) were seperated by a big 99 run gap yet both teams ended with just 69 wins on the season. 

```python
fig15 = px.scatter(Wins, x="Total Wins", y="Total hits",size="Total hits", color="Team", title="Correlation between Wins and Hits Recorded",height=800)
fig15.update_layout(xaxis_title='Wins',yaxis_title='Hits Recorded',showlegend=True)
fig15
```

The relationship between wins and hits is even weaker than that of wins and runs, further hinting at the importance of solid pitching and defense. The Minnesota Twins and the Chicago Cubs, the last and first place team respectively, both had exactly 1,409 hits but the Chicago Cubs still won an outstanding 103 games, outpacing Minnesota, who only had 59 wins, by 44. The Colorado Rockies also out-hit every team but one in the entire league but was still bested by 18 different ballclubs in the overall wins column.
