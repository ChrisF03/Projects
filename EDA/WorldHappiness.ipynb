# Exploratory Data Analysis

## Topic: World Happiness Report (2019 - 2022)

### Abstract

The World Happiness Report is a yearly landmark survey of the state of global happiness. A publication of the United Nations Sustainable Development Solutions Network, it contains articles and rankings of national happiness, based on respondents ratings of their own lives which the report also correlates with various (quality of) life factors.
The World Happiness Report tracks six factors to measure global happiness — healthy life expectancy, economics, social support, absence of corruption, freedom, and generosity.<br><br>
Using reports from the year 2019-2022, I will attempt to answer the following questions when performing my analysis on this dataset:<br><br>
What are the most important contributing factors to a nation’s happiness?<br>
Which country is the happiest in the world?<br>
Which region is the happiest in the world ?<br>
And, finally, has overall happiness increased or decreased since 2019 ?

# World Happiness 2019

```python
df1.set_index('Country',inplace=True)
df1
```

Upon looking at the dataframe, we can see that the country Finland is/was the happiest place on earth in 2019 with an average happiness score of 7.769/10 amongst its citizens. Finland is followed at the top of the list, by its fellow nordic neighbors, Denmark, and Norway, located in northern-northwestern Europe and placing second and third place respectively. At the very bottom of the list, representing the unhappiest place on earth at the same time is the African nation of South Sudan with a low score of 2.853 amongst citizens, slightly beat out by the bordering nation, Central African Republic, and the South Asian nation of Afghanistan. 

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df1, x='Score',kde=True,label=2019,bins=[2,3,4,5,6,7,8])
plt.legend()
plt.show()
```

Here we can see that the majority of the countries rate their happiness between 4 and 6, so about average. The happiest countries rank between 7 and 8. Below we'll take a deeper look at how whow much effect each variable has on overall score. 

### Correlation between variables and score

```python
y = df1[['Score']]
x1 = df1[['GDP per capita']]
x2 = df1[['Social support']]
x3 = df1[['Healthy life expectancy']]
x4 = df1[['Freedom to make life choices']]
x5 = df1[['Generosity']]
x6 = df1[['Perceptions of corruption']]
fig, axis = plt.subplots(2,3, figsize = (15,6))
fig.text(0,0.5,'Happiness Scores',style='normal',fontsize=22,ha='center', va='center', rotation='vertical')
axis[0,0].scatter(x1, y)
axis[0,0].set_xlabel('GDP per Capita',fontsize=14)
axis[0,1].scatter(x2, y)
axis[0,1].set_xlabel('Social support',fontsize=14)
axis[0,2].scatter(x3, y)
axis[0,2].set_xlabel('Healthy life expectancy',fontsize=14)
axis[1,0].scatter(x4, y) 
axis[1,0].set_xlabel('Freedom to make choices',fontsize=14)
axis[1,1].scatter(x5, y)
axis[1,1].set_xlabel('Generosity',fontsize=14)
axis[1,2].scatter(x6, y)
axis[1,2].set_xlabel('Perceived Corruption',fontsize=14)
fig.tight_layout() 
```

Looking at these plots, we can see the correlation between each variable and the world happiness scores in 2019. GDP per capita, a financial metric that breaks down a country's economic output per person and is calculated by dividing the GDP of a nation by its population, has a clear and direct positive relationship with corresponding scores. However, somewhat surprisingly, Social support (defined by the report as having someone to count on in times of trouble) seems to have the strongest correlation to overall feelings of happiness amongst the worlds citizens as it has the steepest slope. The second strongest correlation is Healthy Life Expectancy. Also interestingly, generosity seems to have the least effect on happiness either way, with perceived corruption having the second least effect.

### Happiness by Region

Since we have regional indicator for each country, lets take a look at how many countries are represented in each region and which region is the happiest.

```python
n = (pd.unique(df2['Regional indicator']))
print('The following',len(n),'regions are represented:','\n')
for c in n:
    print(c,'\n')
```

```python
plt.figure(figsize=(8,5))
ri = sns.countplot(y="Regional indicator",data=df2)
plt.show()
s= df2[['Regional indicator']].value_counts() 
s
```

By far, the highest number of countries in the report are located in Sub-Saharan Africa with 39 countries listed, followed by Western Europe and the Latin American and Caribbean region, each with 21.<br><br> Now, which region is the happiest  ? 

```python
Regions = df2[['Regional indicator','Score']]
Regions
```

```python
group_names=['Low','Medium','High']
bin = pd.cut(Regions['Score'], 3, labels=group_names)
Regions['Score-bin'] = bin
Regions
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=Regions, x='Score',kde=True,label=2020,bins=[2,3,4,5,6,7,8], color='green')
plt.legend()
plt.show()
```

Here we can see the score distribution. 

```python
plt.figure(figsize=(15,10))
r = sns.displot(data=Regions, x='Score-bin', hue='Regional indicator')
plt.show()
```

Here we can see the score distribution by region. The unhappiest region in the world is, by far, Sub-Saharan Africa. The whole European continent takes the cake as the happiest region, with the happiest countries in particular being located in Western Europe followed by their neighbors in Central and Eastern Europe.

# Year-to-Year Comparisons

Now we will take a quick, closer look at how World Happiness varied year-to-year by checking the distributions of the score for each year side-by-side. 

```python
x1 = df1[['Score']]
x2 = df2[['Score']]
x3 = df3[['Score']]
x4 = df4[['Score']]
fig, axis = plt.subplots(1,4, figsize = (15,7))
axis[0].boxplot(x1,notch=True)
axis[0].set_ylabel('Score',fontsize=14)
axis[0].set_title('2019')
axis[1].boxplot(x2,notch=True)
axis[1].set_title('2020')
axis[2].boxplot(x3,notch=True)
axis[2].set_title('2021')
axis[3].boxplot(x4,notch=True)
axis[3].set_title('2022')
```

Looking at the boxplots, they look almost identical to each other. However, unsurprisingly, the bottom whisker for the year 2020 is the one that goes the lowest. The year 2021 has the lower whisker the highest out of all, and is the only one that ends above 3. I found that somewhat surprising as the world was still dealing with much of the fallout from the pandemic and from being in quarantine for the better part of a year.

```python
median1 = np.median(x1)
median2 = np.median(x2)
median3 = np.median(x3)
median4 = np.median(x4)
print('Avg Score 2019:',median1,'\n','Avg Score 2020:',median2,'\n','Avg Score 2021:',median3,'\n','Avg Score 2022:',median4)
```

The lowest average happiness score, however, was in 2019 and the highest was in 2022. 

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data=df1, x='Score',label=2019,clip=(2.0, 8.0))
sns.kdeplot(data=df2, x='Score',label=2020,clip=(2.0, 8.0))
sns.kdeplot(data=df3, x='Score',label=2021,clip=(2.0, 8.0))
sns.kdeplot(data=df4, x='Score',label=2022,clip=(2.0, 8.0))
plt.legend()
```

Looking at the Kernel Density plot above gives us another look at the information we just gathered about score averages. 2022 has the highest and 2019, the lowest, further confirming that world happiness has only increased since 2019.
